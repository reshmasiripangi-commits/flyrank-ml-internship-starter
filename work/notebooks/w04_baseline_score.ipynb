## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

Rule:

If a content page has high search volume and low competition, it should receive a higher action score because it has a better chance of gaining traffic.

Reason Codes:

HIGH_SEARCH_VOLUME
LOW_COMPETITION
GOOD_OPPORTUNITY

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("../../data/processed/refresh_feature_vector.csv")

def get_reason(row):
    if row["search_volume"] > 50 and row["competition"] < 0.3:
        return "GOOD_OPPORTUNITY"
    elif row["search_volume"] > 50:
        return "HIGH_SEARCH_VOLUME"
    else:
        return "LOW_PRIORITY"

df["reason_code"] = df.apply(get_reason, axis=1)

df[["search_volume","competition","reason_code"]].head()
```

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

Each content item receives a score based on search volume and competition.

Higher search volume increases the score.

Higher competition decreases the score.

The ranked list is sorted from highest score to lowest score.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
df["score"] = df["search_volume"] * (1 - df["competition"])

ranked = df.sort_values("score", ascending=False)

ranked.head(20)
ranked.to_csv("../../outputs/baseline_action_score.csv", index=False)

print("CSV Saved Successfully")
```
